# 修改数组元素

```python
t=np.arange(30).reshape(5,6)
t[(t>5)&(t<20)]=10 # 与
t[(t<6)|(t>25)]=-1 # 或
t[~(t<=19)]=5 # 非
print(t)
```

```python
t=np.where(t==10,True,False)
print(t)
```

```python
# 往哪个轴添加，哪个轴的size就变大
a=np.array([[1,2,3],[4,5,6]])
print('不指定轴，则展开：')
print(np.append(a, [7,8,9]))
print('沿轴0添加元素：')
print(np.append(a, [[7,8,9]],axis=0))
print('沿轴1添加元素：')
print(np.append(a, [[5,5,5], [7,8,9]],axis=1))
```
