# 调整数组形状
#### reshape( )

```python
four=np.array([[1,2,3],[4,5,6]]) # 两行三列
#修改的是原有的
four.shape= (3,2)
print(id(four))

#返回一个新的数组
four2=four.reshape(3,2)
print(id(four2))
```

```python
#默认情况下‘C’以行为主的顺序展开，‘F’（Fortran风格）意味着以列的顺序展开
five=four.flatten(order='C')
six=four.flatten(order='F')
print(five)
print(six)
```

# 数据运算

```python
# ndim相同才可以运算
# ndim不同时，只有常数 或者 其中一个ndarray必须对应的轴的size为1
t1=np.arange(10).reshape(2,5)
t2=np.arange(10,20).reshape(2,5)
print(t1+t2)
print(t1*t2)
```

```python
a = np.array([[1,2,3],[4,5,6]])
# 按哪个轴求和，哪个轴消失
print(np.sum(a,axis=0)) # [5 7 9]
print(np.sum(a,axis=1)) # [ 6 15]
```
