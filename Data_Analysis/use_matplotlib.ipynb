# 绘制折线图

```python
import random
# ----- 1. 准备数据 -----
# x轴：周一到周日（数字1-7）
x = range(1, 8)
# 定义3个城市的温度数据（随机生成20-35℃的数值）
city1_temp = [random.randint(20, 35) for _ in x]  # 城市1温度
city2_temp = [random.randint(18, 32) for _ in x]  # 城市2温度
city3_temp = [random.randint(22, 38) for _ in x]  # 城市3温度

# ----- 2. 设置画布和基础样式 -----
plt.figure(figsize=(12, 6))  # 画布大小
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] # 解决中文显示问题（Mac）

# ----- 3. 设置坐标轴刻度和标签 -----
# x轴：替换为星期几，旋转45度避免重叠
x_labels = ['周一', '周二', '周三', '周四', '周五', '周六', '周日']
plt.xticks(x, x_labels, rotation=45, fontsize=10)

# y轴：温度范围（取所有数据的最小/最大值，保证刻度覆盖全部数据）
all_temp = city1_temp + city2_temp + city3_temp
y_min = min(all_temp) - 2  # 留2℃余量
y_max = max(all_temp) + 2
plt.yticks(range(y_min, y_max + 1), fontsize=10)
plt.ylabel('温度 (℃)', fontsize=12)  # y轴名称

# ----- 4. 绘制多条折线 -----
# 每条线设置不同颜色、线型、标记，方便区分
plt.plot(x, city1_temp,
         color='red',      # 颜色
         linestyle='-',    # 线型（实线）
         linewidth=1,      # 线宽
         marker='o',       # 数据点标记（圆形）
         markersize=10,    # 标记大小
         label='北京')     # 图例标签

plt.plot(x, city2_temp,
         color='blue',
         linestyle='--',   # 线型（虚线）
         linewidth=1,
         marker='s',       # 数据点标记（正方形）
         markersize=10,
         label='上海')

plt.plot(x, city3_temp,
         color='green',
         linestyle='-.',   # 线型（点划线）
         linewidth=1,
         marker='^',       # 数据点标记（三角形）
         markersize=10,
         label='广州')

# ----- 5. 美化和补充信息 -----
plt.title('一周内北京/上海/广州温度对比', fontsize=14, pad=20)
plt.legend(loc='upper right', fontsize=10)  # 显示图例
plt.grid(True, alpha=0.3)  # 显示网格，alpha设置透明度
plt.tight_layout()  # 自动调整布局，防止标签被截断
```

# 绘制散点图

```python
# ----- 1. 准备数据 -----
# 模拟10天的气温（20-35℃）和对应冰淇淋销量（100-500份）
temperature = [random.randint(20, 35) for _ in range(10)]
sales = [temp * 12 + random.randint(20, 80) for temp in temperature]  # 销量与气温正相关

# ----- 2. 基础设置 -----
plt.figure(figsize=(10, 6))
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False  # 解决Mac下负号显示异常问题

# ----- 3. 绘制散点图（核心区别区） -----
# 散点图用plt.scatter()
plt.scatter(
    temperature,  # x轴数据（气温）
    sales,        # y轴数据（销量）
    s=80,         # 点的大小（size），折线图无此参数
    c='orange',   # 点的颜色
    marker='o',   # 点的形状（圆形）
    alpha=0.7,    # 点的透明度（0-1），折线图可设置但少用
    edgecolors='black',  # 点的边框颜色，折线图无此参数
    label='每日销量'
)

# ----- 4. 坐标轴与美化 -----
plt.xlabel('气温 (℃)', fontsize=12)
plt.ylabel('冰淇淋销量 (份)', fontsize=12)
plt.title('气温与冰淇淋销量关系散点图', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
```
