```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,100)
fig=plt.figure(figsize=(20,10))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

ax1.plot(x,x)
ax2.plot(x,-x)
ax3.plot(x,x**2)
ax3.grid(color='yellow',linestyle='--',linewidth=1,alpha=0.5)
ax4.plot(x,np.log(x))
```

```python
list1= [1,2,3,4]
oneArray=np.array(list1)
print(type(oneArray))
print(oneArray)
```

```python
#创建数组的多种形式
# 1.直接传入列表的方式
t1=np.array([1,2,3])
print(t1)
print(type(t1))
print('-'*50)
# 2.传入range生成序列
t2=np.array(range(10))
print(t2)
print(type(t2))
print('-'*50)
# 3.使用numpy自带的np.arange()生成数组
t3=np.arange(0,10,2)
print(t3)
print(type(t3))
```

```python
my_sum=np.sum(t3)
print(my_sum)
```

# 常用属性

```python
list2= [[1,2],[3,4],[5,6]]
twoArray=np.array(list2)

print(twoArray.ndim) # 获取数组的维度(注意: 与函数的参数很像)
print(twoArray.shape) # 形状（行，列）
print(twoArray.size) # 有多少个元素
print(twoArray.dtype) # 数据类型
```

# 调整数组形状
#### reshape( )

```python
four=np.array([[1,2,3],[4,5,6]]) # 两行三列
#修改的是原有的
four.shape= (3,2)
print(id(four))

#返回一个新的数组
four2=four.reshape(3,2)
print(id(four2))
```

```python
#默认情况下‘C’以行为主的顺序展开，‘F’（Fortran风格）意味着以列的顺序展开
five=four.flatten(order='C')
six=four.flatten(order='F')
print(five)
print(six)
```

```python
#拓展：数组的形状
t=np.arange(24)
print(t)
print(t.shape)

#转换成二维
print('-'*50)
t1=t.reshape((4,6))
print(t1)
print(t1.shape)
```
